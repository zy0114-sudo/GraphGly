```python
#B_positive

def get_biaowei(pdb_file_path,residue):
    B_line=[]
    A_line=[]
    B_count=0
    A_count=0
    with open(pdb_file_path,'r') as file:
        line = file.readline()    
        B_list=[]
        A_list=[]      
        while line:
            dat_in = line[0:80].split()
            if (dat_in[0] == 'ATOM'):
                chain_id = line[21]
                #residue_id = int((line[22:26]))  
                x = float(line[30:38])
                y = float(line[38:46])
                z = float(line[46:54])
                atom_type2=line[77]
                residue_type = line[17:20]
                residue_id2 = str(line[22:27]).strip()  
                    #print(residue_type)
                    # First try CA distance of contact map
                atom_type = line[13:16].strip()
                if chain_id=='B':   #蛋白B
                    if residue_type in residue:
                        if atom_type2 != 'H' :
                            B_list.append([x,y,z,atom_type,B_count,residue_id2,residue_type,chain_id,atom_type2])
                            B_line.append(line)
                            B_count += 1 
                elif chain_id=='A':
                    if residue_type in residue:
                        if atom_type2 !='H':
                            A_list.append([x,y,z,atom_type,A_count,residue_id2,residue_type,chain_id,atom_type2])
                            A_line.append(line)
                            A_count += 1
            elif line.startswith('TER'):
                pass
            line = file.readline()
    return(B_list,B_line,A_list,A_line,B_count,A_count)    
def compute_B(B_list,A_list,A_count):
    #获取B链对A的表位
    atom_len1=len(B_list)  #蛋白B
    atom_len2=len(A_list)  #蛋白A 
    cut_off=5
    cut_off=cut_off**2
    rog1=[]
    if A_count!=0:
        for i in range(atom_len1):  #蛋白B
            atom1=B_list[i]
            ab_chain=atom1[7]
            #ab_atom_name=atom1[3]
            ab_residue_id=atom1[5]
            #ab_residue_type=atom1[6]
            #ab_atom=atom1[8]
            
            for j in range(atom_len2):  #蛋白A
                atom2=A_list[j]
                distance=0
                for k in range(3):
                    distance+=(atom1[k]-atom2[k])**2
                if distance<=cut_off:
                    rog1.append((ab_chain,ab_residue_id))              
    return rog1

def From_interface(rog,B_line,path_r,pdb2):
    final_receptor=[]
    for i in B_line:
        chain_id=i[21]
        residue_id2 = str(i[22:27]).strip()  
        for j in rog:  
            ch=j[0]
            re=j[1]
            if chain_id == ch and residue_id2 == re:
                final_receptor.append(i)
    Write_Interface(final_receptor,path_r,"%s_B.pdb"%(pdb2))

def Write_Interface(final_atom,path,name):
    if not final_atom:
        return None
    new_path=os.path.join(path,name)
    with open(new_path,'w') as file:
        for line in final_atom:
            file.write(line)   
import pandas as pd 
import os 
data=pd.read_csv('/data5_large/home/yanzh/deeplearn/graph_residue/my_data/protein_new_3222/dataset_shuanglian_buchong/proA/protein_all/file_list.csv')
residue= ['ALA', 'CYS', 'ASP', 'GLU', 'PHE', 'GLY', 'HIS', 'ILE', 'LYS', 'LEU', 'MET', 'ASN', 'PRO', 'GLN', 'ARG', 'SER', 'THR', 'VAL', 'TRP', 'TYR']

for index,value in data.iterrows():
    pdb=value['pdb']
    print(pdb)
    pdb2=pdb[0:4]
    path_r='/data5_large/home/yanzh/deeplearn/graph_residue/my_data/protein_new_3222/dataset_shuanglian_buchong/proB/protein_postive'
    pdb_file_path=f'/data5_large/home/yanzh/deeplearn/graph_residue/my_data/protein_new_3222/dataset_shuanglian_buchong/chain-all/{pdb2}.ent.pdb'
    B_list,B_line,A_list,A_line,B_count,A_count=get_biaowei(pdb_file_path,residue)
    rog=compute_B(B_list,A_list,A_count)
    rog=list(set(rog))
    From_interface(rog,B_line,path_r,pdb2)
```
